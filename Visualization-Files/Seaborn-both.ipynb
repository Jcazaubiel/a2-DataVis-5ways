```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Load and format dataset
cars = pd.read_csv('cars-sample.csv')
cars.drop(cars.columns[[0]], axis=1, inplace = True)


# Create color palette
colors = ['#FF0000', '#CC9900', '#00CC00', '#0066FF', '#CC0099']

# Create bubble plot
graph = sns.scatterplot(data=cars, x = 'Weight', y = 'MPG', size = 'Weight', 
                        hue = 'Manufacturer', palette = colors, alpha = 0.5, sizes = (10, 200))


# Set the gridlines and ticks for the X axis
graph.set_xticks([2000, 3000, 4000, 5000])
graph.set_xticks([1500, 2500, 3500, 4500], minor=True)

# Set the gridlines and ticks for the Y axis
graph.set_yticks([10, 20, 30, 40])
graph.set_yticks([15, 25, 35, 45], minor=True)

# Set the parameters for the major/minor gridlines
graph.grid(b=True, which='major', linewidth=1.25)
graph.grid(b=True, which='minor', linewidth=0.75)

# Set the style of the graph
sns.set_style('darkgrid')

# Set the style of the ticks
graph.tick_params(direction = 'out', color = 'black', bottom = True, left = True)


# Set the axis labels
plt.xlabel('Weight', weight = 'bold')
plt.ylabel('MPG', weight = 'bold')

# Move the legend to the right of the graph
plt.legend(bbox_to_anchor=(1, 1), fontsize=10)

# Show plot
plt.show()
```
